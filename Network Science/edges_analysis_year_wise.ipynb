```python
year = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']

usa_length = []

excel = pd.read_csv('C:/Users/Krishan/Downloads/FT50_WOS/insitutional_year_wise_all/edges/edges_2005.csv')    
insitute_2005 = excel.loc[(excel['Source'].apply(lambda x: x[-4:] == 'USA.')) & (excel['Target'].apply(lambda x: x[-4:] == 'USA.'))]
insitute_2005 = insitute_2005.sort_values(by='Weight', ascending=False)
insitute_2005 = insitute_2005.groupby(['Source']).agg({'Source':'count', 'Weight':'sum'}).sort_values(by='Weight', ascending=False)
insitute_2005.rename(columns={'Source': 'No of Insititute'}, inplace=True)
insitute_2005.reset_index(level=0, inplace=True)
 

for i in range(0, len(year)):    
    excel = pd.read_csv('C:/Users/Krishan/Downloads/FT50_WOS/insitutional_year_wise_all/edges/edges_'+str(year[i])+'.csv')    
    insitute_with_in_usa = excel.loc[(excel['Source'].apply(lambda x: x[-4:] == 'USA.')) & (excel['Target'].apply(lambda x: x[-4:] == 'USA.'))]
    insitute_with_in_usa = insitute_with_in_usa.groupby(['Source']).agg({'Source':'count', 'Weight':'sum'}).sort_values(by='Weight', ascending=False)
    insitute_with_in_usa.rename(columns={'Source': 'No of Insititute'}, inplace=True)
    insitute_with_in_usa.reset_index(level=0, inplace=True)
    insitute_2005 = pd.merge(insitute_2005, insitute_with_in_usa, on='Source', how='outer', suffixes=('2005' if i ==0 else year[i-1], year[i]))
insitute_2005 = insitute_2005.sort_values(by='No of Insititute2005', ascending=False)
all_sum = insitute_2005['No of Insititute2005'] + insitute_2005['No of Insititute2006']+insitute_2005['No of Insititute2007']+insitute_2005['No of Insititute2008']+insitute_2005['No of Insititute2009']+insitute_2005['No of Insititute2010']+insitute_2005['No of Insititute2011']+insitute_2005['No of Insititute2012']+insitute_2005['No of Insititute2013']+insitute_2005['No of Insititute2014']+insitute_2005['No of Insititute2015']+insitute_2005['No of Insititute2016']+insitute_2005['No of Insititute']
insitute_2005['all_insitute'] = all_sum
```
