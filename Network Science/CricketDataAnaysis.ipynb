```python
ODI_Match_Results['Start Date'] = ODI_Match_Results['Start Date'].apply(dateutil.parser.parse, dayfirst=True)
ODI_Match_Results
```

```python
ground_name = ["Lord's", "Southampton", "Nottingham", "Cardiff", "The Oval", "Birmingham", "Manchester", "Chester-le-Street", "Leeds", "Bristol"]
```

```python
date = parse('2012-01-01')
date
```

```python
our_result = ODI_Match_Results.loc[ODI_Match_Results["Ground"].apply(lambda x: x in ground_name) & ODI_Match_Results["Start Date"].apply(lambda x: x > date)]
 
our_result1 = our_result[our_result['Result']== 'won'].groupby(['Country'])['Result'].count()
our_result2 = our_result[our_result['Result']== 'lost'].groupby(['Opposition'])['Result'].count()
```

```python
contry = our_result2.index
won_array = our_result1.values
loss_array = our_result2.values
com_lss_win =  won_array + loss_array
print(com_lss_win)
print(contry)
```
