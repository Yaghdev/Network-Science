```python
year = ['2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
ary = ['Accounting','Economics','Economics, Management', 'Entrepreneurship', 'Ethics', 'Finance', 'Human Resources', 'International Business', 'Management','Marketing', 'Operations & Information Systems', 'Organizational Behaviour']

for area in year:    
    excel = pd.read_csv('D:/new_ft50/edges_excel/insitutional_edege'+str(area)+'.csv')
    insitute_with_in_usa = excel.loc[(excel['Source'].apply(lambda x: x[-4:] == 'USA.')) & (excel['Target'].apply(lambda x: x[-4:] != 'USA.')) | (excel['Source'].apply(lambda x: x[-4:] != 'USA.')) & (excel['Target'].apply(lambda x: x[-4:] == 'USA.'))]
    insitute_with_in_usa = insitute_with_in_usa.sort_values(by='Weight', ascending=False)
    result = insitute_with_in_usa.groupby(['Source']).agg({'Source':'count', 'Weight':'sum'}).sort_values(by='Weight', ascending=False)
    result['rank'] = [x for x in range(1, len(result)+1)]
    result['percentile'] = (len(result)-result['rank'])/len(result)*100
    result['quantiles'] = pd.qcut(result['rank'], 4, labels=False)
    result.loc[result['quantiles'] == 0] 
    print(result)
    break
                                                            
```
