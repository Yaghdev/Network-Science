```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="darkgrid")
```

```python
usa_length = []
nonusa_length = []
year = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
for x in year:
    insitute = pd.read_excel('D:/new_ft50/Insitutional_result/year_wise/Insitutional_network_'+str(x)+'.xlsx')  
    insitute = insitute.sort_values(by='degree', ascending=False)   
    insitute_usa = insitute.loc[insitute['Insitute'].apply(lambda x: x[-4:] == 'USA.')]
    insitute_nonusa = insitute.loc[insitute['Insitute'].apply(lambda x: x[-4:] != 'USA.')]
    nonusa_length.append(len(insitute_nonusa))
    usa_length.append(len(insitute_usa))
    
usa_length = np.array(usa_length)
non_usa_length = np.array(nonusa_length)
usa_percentage = (usa_length/(usa_length + non_usa_length))*100
```

```python
t = np.array(year)
data1 = np.array(usa_length)
data2 = np.array(nonusa_length)
data3 = np.array(usa_percentage)
fig, ax1 = plt.subplots(figsize=(10, 8))

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('number of institute', color=color)
ax1.bar(t-0.2, data1,  width=0.40, color='0.7',align='center', label="dsvsfv")
ax1.bar(t+0.2, data2, width=0.40,color='r',align='center',  label="fcsdfgr")
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()   

color = 'tab:blue'
ax2.set_ylabel('USA Percentage', color=color)
ax2.plot(t, data3, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()   
plt.show()
```
